# EDA — NYC 311 + Weather

```
import pandas as pd, matplotlib.pyplot as plt
df = pd.read_csv('data/final/nyc_311_weather_final.csv', parse_dates=['created_date','closed_date'])
df.head()
```

```
daily = df.groupby('date').size()
daily.plot(kind='line', figsize=(12,4), title='Incidencias diarias'); plt.show()
```
